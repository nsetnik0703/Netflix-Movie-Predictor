# Scrape the movie budges


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from sqlalchemy import create_engine
```

```python
# Define a function to scrape the movies
def scrape_movies(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text.encode('ascii', 'ignore').decode('ascii'), 'html.parser')
    df = []
    for tr in soup.find_all('tr')[1:]:
        tds = tr.find_all('td')
        data = [td.text for td in tds]
        df.append(data) 
    return df
```

```python
link = []
for i in range(6):
    for j in range (10):
        if i == 0:
            if j == 0:
                link.append('/all')
            else:
                link.append('/all/'+str(j)+'01')
        else: 
            link.append('/all/'+str(i)+str(j)+'01')
            
```

```python
for i in range(len(link)):
    print(str(i)+': '+link[i])
```

```python
len(link)
```

```python
result = scrape_movies('https://www.the-numbers.com/movie/budgets'+link[0])
for n in range(1, len(link)): 
    df = scrape_movies('https://www.the-numbers.com/movie/budgets'+link[n])
    if len(df) > 0:
        result += df
```

```python
movies_df = pd.DataFrame(result, columns=['Number', 'Release Date', 'Movie', 'Production Budget', 'Domestic Gross', 'Worldwide Gross'])
movies_df
```

```python
movies_df.dtypes
```

```python
movies_df['Release Date'] = pd.to_datetime(movies_df['Release Date'])
```

```python
# Define a function to parse the money
def parse_money(row):
    money = row.replace("$", "")
    money = money.replace(",", "")
    return float(money)
```

```python
movies_df['Production Budget'] = movies_df['Production Budget'].apply(parse_money)
movies_df['Domestic Gross'] = movies_df['Domestic Gross'].apply(parse_money)
movies_df['Worldwide Gross'] = movies_df['Worldwide Gross'].apply(parse_money)
```
